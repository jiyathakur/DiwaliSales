STATE

```python
plt.figure(figsize=(12,9))
plt.xticks(rotation=75)
ax=sns.countplot(x=sales_df.State)
for bars in ax.containers :
    ax.bar_label(bars)
```

no of customers from corressponding states.max no of customers are from UP.

```python
#order from top 10 states
ok=sales_df.groupby('State',as_index=False)['Orders'].sum()
ok
```

```python
plt.figure(figsize=(12,6))
sns.set_style('darkgrid')
sns.barplot(x='State',y='Orders',data=ok)
plt.xticks(rotation=78);
```

so max orders were placed from UP followed by Maharashtra.

```python
ax=sns.countplot(x=sales_df.Shaadi)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df=sales_df.groupby(['Shaadi','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
df
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Shaadi',y='Amount',hue="Gender",data=df)
```

most buyers are married females.

```python
plt.figure(figsize=(12,6))
plt.title('total no. of customers corresponding to their occupation')
sns.countplot(x=sales_df.Occupation)
plt.xticks(rotation=75);
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
ax=sns.countplot(x='Occupation',data=sales_df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df=sales_df.groupby(['Occupation','Gender'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
sns.barplot(x='Occupation',y='Amount',hue='Gender',data=df)
```

most of the buyers are working in IT sector followed by Aviation.

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
ax=sns.countplot(x=sales_df.Product_Category)
for bars in ax.containers :
    ax.bar_label(bars)
```

```python
df=sales_df.groupby(['Product_Category','Gender'],as_index=False)['Orders'].sum().sort_values('Orders',ascending=False)
df
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
sns.barplot(x='Product_Category',y='Orders',hue='Gender',data=df)
```

Clothing and Apparel has more orders followed by Food.

```python
#top 10 most sold product with product_id
df=sales_df.groupby(['Product_ID'],as_index=False)['Orders'].sum().head(10).sort_values('Orders',ascending=False)
plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
sns.barplot(x='Product_ID',y='Orders',data=df)
```

CONCLUSION
Females having age group between 26-35 shop more are from UP,Maharashtra,Karnataka.
